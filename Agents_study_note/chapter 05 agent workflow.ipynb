# 05 低代码的智能体搭建

## 总体收获

这一次学的是低代码平台里的智能体构件。低代码平台最开始接触的是 Dify，然后我也注意到 Coze 最近很流行，就顺便看了一下相关教程，但整体看下来，我觉得 Coze 好像并不是很适合我。

实际上我看了好几个低代码平台的介绍之后，反而觉得真正有帮助的，是对这些平台优点和缺点的分析。因为 Dify 有学生账号，可以免费使用一年的 Pro 版，所以如果是作业层面的需求，我可能还是会尝试用 Dify。但说实话，在看了 Dify 的一些应用案例之后，那种“食之无味，弃之可惜”的感觉依然非常强。至少对于个人来说如此，对于企业我觉得还是有很多可以利用的地方。

教程里反复强调低代码平台的优势在于对技术要求低，但在我看来，即便只是涉及 API、JSON 这类数据格式，其实已经足够把很多非技术用户挡在门外了。因为我自己也是自学过来的，很清楚这些概念对外行来说有多难理解，远远谈不上真正意义上的低门槛。

Dify 的可视化设计，如果前提是你已经理解了基本的数据结构和节点的概念，那确实是可以上手的。但它给出的那些案例，我依然觉得属于给人感觉“哎呀……emmmm”的那一类。

比如案例里个人助手所提供的几个功能，整合在一个 Agent 里并没有太大的实际意义。视频生成我本来就有自己常用的平台，图片也是，而且这些工具的难点是prompt调试，并不是进入哪个工作流。

如果只是问天气，打开手机自带的天气预报，甚至问下Siri，速度远远快过还要去问一个 Agent 今天天气怎么样。这些功能更多还是大家在想象中的“生活助理”，而不是现实中真正会使用的需求。

至于推荐菜谱，会做饭的人（例如我）不太会为这个纠结，更不会想让 AI 来决定今天做什么，毕竟做菜讲究当时当地，季节，菜价波动，人的身体状况等等……这些因素 AI 都无法考虑到……感觉只有云做饭的人才想得出这需求。

不会做饭的人直接点外卖，这类需求本身就比较虚。

最后说回 Dify。我觉得它真正可能被用到的，还是RAG这一类应用。但问题在于，我自己是在本地电脑上跑 RAG，用的是 OpenAI 的 embedding。如果用 Copilot 来为自己定制一个程序，虽然过程会麻烦一点，但因为我是全程用自然语言和 Copilot 交互，并不需要自己写代码，最终得到的定制化效果反而比 Dify 的工作流好得多。尤其是 Dify 目前的切片方式，导致很多我真正需要的功能做不了，比如从论文中稳定地提取 reference、并且保留 reference 之间的关联关系。这些对我来说是核心需求，但它并不能很好地支持。
