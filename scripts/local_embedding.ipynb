```python
from sentence_transformers import SentenceTransformer

# 多语言（推荐）
model = SentenceTransformer(
    "Alibaba-NLP/gte-multilingual-base",
    trust_remote_code=True
)

# 示例句子
sentences = ["这是一句中文测试", "This is an English test."]
embeddings = model.encode(sentences, convert_to_tensor=True)

print(embeddings.shape)  # 输出 (2, 768)
```

```python
import time
# 记录开始时间
start_time = time.time()

# 示例句子
sentences = ["这是一句中文测试", "Orientation to the Canadian Health Care System Part II focuses on the provincial delivery of health care including the importance of patient-centered care, safe, competent and ethical practice, inter-professional collaboration, information on levels of care and regulated health professions."]
embeddings = model.encode(sentences, convert_to_tensor=True)

# 记录结束时间
end_time = time.time()

print(embeddings.shape)  # 输出 (2, 768)
print(f"编码时间: {end_time - start_time} 秒")
```
